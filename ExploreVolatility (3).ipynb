```python
#organize dates in order --> need all 3 values on same day to calculate
v_dates = v.loc[:, 'Trade Date'].values
vmin = np.amin(v_dates)
vv_dates = vv.loc[:, 'Trade Date'].values
vvmin = np.amin(vv_dates)
spy_dates = spy.loc[:, 'Date'].values
spymin = np.amin(v_dates)

start_date = np.argmax([vmin, vvmin, spymin]) #first date will all 3 values
dates = np.flip([v_dates, vv_dates, spy_dates][start_date], axis = 0)
dates #ordered dates for comparison
```

```python
#organize values into new array (change to dataframe?)
daily_values = np.empty((len(dates), 11), dtype = object)

for i in range(len(dates)):
    v_close = v.loc[v.loc[:, 'Trade Date'] == dates[i], 'Close'].values[0]
    vv_close = vv.loc[vv.loc[:, 'Trade Date'] == dates[i], 'Close'].values[0]
    spy_close = spy.loc[spy.loc[:, 'Date'] == dates[i], 'Close'].values[0]
    daily_values[i, 0] = dates[i]
    daily_values[i, 1] = v_close
    daily_values[i, 2] = vv_close
    daily_values[i, 3] = spy_close
    daily_values[i, 4] = (vv_close - v_close)/vv_close
    
summary = pd.DataFrame(daily_values, columns = ['Date', 'V', 'VV', 'SPY', 'VV - V / VV', 'Holding',\
                                                 'Held Value', 'Cash', 'Trade Sum', 'Long Term Shares', \
                                                'Long Term Value'], copy = True)
```

Define the initial amount of money invested and the cost per trade.
Can add additional inputs if needed.
Dates range from 11/17/2005 to 2/12/2018

```python
initial_investment = 1000 #amount of money to invest initially (assumes only 1 influx of money)
trade_cost = 5 #fixed amount, can change to percentage
threshold = 0.1 #to make decision set at a threshold above or below 0 (VV - V / VV)
```

```python
#summary['VV - V / VV'] = (summary['VV'] - summary['V']) / summary['VV']

def simulate_returns(dataframe, initial_investment, trade_cost, threshold):
    """
    Simulates returns using existing dataset (fixed dataset based on summary dataframe)
    allows for varying the initial_investment, trade_cost, and threshold
    """
    
    summary = dataframe.copy()
    
    summary.iloc[0, 5:] = 0 #initialize as 0 holdings
    LTshares = (initial_investment - trade_cost) / summary.loc[0, 'SPY']
    summary.loc[0, 'Long Term Shares'] = LTshares
    summary.loc[0, 'Long Term Value'] = initial_investment - trade_cost
    summary.loc[0, 'Cash'] = initial_investment

    def update_row(df, row):
        """
        basic steps that will be repeated to update each row of summary DF
        """
        df.loc[row, 'Held Value'] = df.loc[row, 'Holding']*df.loc[row, 'SPY']
        df.loc[row, 'Long Term Shares'] = LTshares
        df.loc[row, 'Long Term Value'] = df.loc[row, 'Long Term Shares']*df.loc[row, 'SPY']
        df.loc[row, 'Trade Sum'] = df.loc[row, 'Cash'] + df.loc[row, 'Held Value']
        return df

    for i in range(1, summary.shape[0]):
        #print(i)
        if np.absolute(summary.loc[i, 'VV - V / VV']) <= threshold: #do nothing
            #print('do nothing')
            summary.loc[i, 'Holding'] = summary.loc[i-1, 'Holding']
            summary.loc[i, 'Cash'] = summary.loc[i-1, 'Cash']
            summary = update_row(summary, i)

        elif (summary.loc[i, 'VV - V / VV'] + threshold > 0) & (summary.loc[i-1, 'Holding'] == 0): #buy
            #print(buy)
            summary.loc[i, 'Holding'] = (summary.loc[i-1, 'Cash'] - trade_cost) / summary.loc[i, 'SPY']
            summary.loc[i, 'Cash'] = 0
            summary = update_row(summary, i)

        elif (summary.loc[i, 'VV - V / VV'] - threshold < 0) & (summary.loc[i-1, 'Held Value'] != 0): #sell
            #print('sell')
            summary.loc[i, 'Cash'] = summary.loc[i-1, 'Held Value'] - trade_cost
            summary.loc[i, 'Holding'] = 0
            summary.loc[i, 'Held Value'] = 0
            summary = update_row(summary, i)

        else: #also do nothing
            #print('do nothing')
            summary.loc[i, 'Holding'] = summary.loc[i-1, 'Holding']
            summary.loc[i, 'Cash'] = summary.loc[i-1, 'Cash']
            summary = update_row(summary, i)
            
    return summary
```

Run simulations for multiple costs per trade and thresholds:

```python
summary1 = simulate_returns(summary, 1000, 0, 0)

fig = plt.figure(figsize = (12, 12))
plt.plot(np.arange(summary1.shape[0]-1), summary1.loc[1:, 'Trade Sum'], label = 'Active Trading')
plt.plot(np.arange(summary1.shape[0]-1), summary1.loc[1:, 'Long Term Value'], label = 'Buy and Hold SPY')
plt.title('Cost per Trade = 0, Trade threshold = 0')
plt.legend()
```

```python
summary1 = simulate_returns(summary, 1000, 5, 0)

fig = plt.figure(figsize = (12, 12))
plt.plot(np.arange(summary1.shape[0]-1), summary1.loc[1:, 'Trade Sum'], label = 'Active Trading')
plt.plot(np.arange(summary1.shape[0]-1), summary1.loc[1:, 'Long Term Value'], label = 'Buy and Hold SPY')
plt.title('Cost per Trade = 5, Trade threshold = 0')
plt.legend()
```
