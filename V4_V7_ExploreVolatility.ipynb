```python
#organize dates in order --> need all values on same day to calculate
v_dates = v.loc[:, 'Trade Date'].values
vmin, vmax = np.amin(v_dates), np.amax(v_dates)
vv_dates = vv.loc[:, 'Trade Date'].values
vvmin, vvmax = np.amin(vv_dates), np.amax(vv_dates)
spy_dates = spy.loc[:, 'Date'].values
spymin, spymax = np.amin(spy_dates), np.amax(spy_dates)
vxx_dates = vxx.loc[:, 'Date'].values
vxxmin = np.amin(vxx_dates)
v4_dates = v4.loc[:, 'Trade Date'].values
v4min = np.amin(v4_dates)
v7_dates = v7.loc[:, 'Trade Date'].values
v7min = np.amin(v7_dates)

start_date = np.argmax([vmin, vvmin, spymin, vxxmin, v4min, v7min]) #first date will all values
dates = [v_dates, vv_dates, spy_dates, vxx_dates, v4_dates, v7_dates][start_date]
dates #= np.flip(dates, axis = 0) #ordered dates for comparison
```

```python
#organize values into new array / dataframe
columns = ['Date', 'V', 'VV', 'SPY', 'VXX', 'V4', 'V7', '(VV - V) / VV',\
           '[((VV-V)/VV) - ((V7-V4)/(V7*3))]', 'Holding',\
           'Held Value', 'Cash', 'Trade Sum', 'Long Term Shares', 'Long Term Value']
                                                

daily_values = np.empty((len(dates), len(columns)), dtype = object)

for i in range(len(dates)):
    if dates[i] not in v['Trade Date'].values:
        #daily_values[i, 0] = np.nan
        continue
    v_close = v.loc[v.loc[:, 'Trade Date'] == dates[i], 'Close'].values[0]
    vv_close = vv.loc[vv.loc[:, 'Trade Date'] == dates[i], 'Close'].values[0]
    spy_close = spy.loc[spy.loc[:, 'Date'] == dates[i], 'Close'].values[0]
    vxx_close = vxx.loc[vxx.loc[:, 'Date'] == dates[i], 'Close'].values[0]
    v4_close = v4.loc[v4.loc[:, 'Trade Date'] == dates[i], 'Close'].values[0]
    v7_close = v7.loc[v7.loc[:, 'Trade Date'] == dates[i], 'Close'].values[0]
    daily_values[i, 0] = dates[i]
    daily_values[i, 1] = v_close
    daily_values[i, 2] = vv_close
    daily_values[i, 3] = spy_close
    daily_values[i, 4] = vxx_close
    daily_values[i, 5] = v4_close
    daily_values[i, 6] = v7_close
    daily_values[i, 7] = (vv_close - v_close)/vv_close
    daily_values[i, 8] = daily_values[i, 7] - ((v7_close - v4_close)/(v7_close * 3))
    
summary = pd.DataFrame(daily_values, columns = columns, copy = True)
summary.dropna(axis = 0, how = 'all', inplace = True)
summary.reset_index(drop = True, inplace = True)
summary.head()
```

Define the initial amount of money invested and the cost per trade.
Can add additional inputs if needed.
Dates range from 2009 to 2018

```python
initial_investment = 1000 #amount of money to invest initially (assumes only 1 influx of money)
trade_cost = 0 #fixed amount, can change to percentage
threshold = 0 #to make decision set at a threshold above or below 0 (VV - V / VV)
```

```python
#if ['VV - V / VV'] > 0 , buy SPY. If < 0, sell SPY

def simulate_returnsVV(dataframe, initial_investment, trade_cost, threshold, start_date=None):
    """
    Simulates returns from shorting VXX using existing dataset (fixed dataset based on summary dataframe)
    allows for varying the initial_investment, trade_cost, threshold, and start date (date must be in dataframe)
    Utilizes old formula (VV-V)/VV
    """
    
    if start_date is not None:
        d1 = np.where(dataframe['Date'].values == start_date)[0][0]
        summary = dataframe.iloc[d1:, :].copy()
        summary.reset_index(drop = True, inplace = True)
    else:
        summary = dataframe.copy()
    
    summary = dataframe.copy()
    
    summary.loc[0, 'Holding':] = 0 #initialize as 0 holdings
    LTshares = (initial_investment - trade_cost) / summary.loc[0, 'SPY']
    summary.loc[0, 'Long Term Shares'] = LTshares
    summary.loc[0, 'Long Term Value'] = initial_investment - trade_cost
    summary.loc[0, 'Cash'] = initial_investment

    def update_row(df, row):
        """
        basic steps that will be repeated to update each row of summary DF
        """
        df.loc[row, 'Held Value'] = df.loc[row, 'Holding']*df.loc[row, 'SPY']
        df.loc[row, 'Long Term Shares'] = LTshares
        df.loc[row, 'Long Term Value'] = df.loc[row, 'Long Term Shares']*df.loc[row, 'SPY']
        df.loc[row, 'Trade Sum'] = df.loc[row, 'Cash'] + df.loc[row, 'Held Value']
        return df

    for i in range(1, summary.shape[0]):
        #print(i)
        if np.absolute(summary.loc[i, '(VV - V) / VV']) <= threshold: #do nothing
            #print('do nothing')
            summary.loc[i, 'Holding'] = summary.loc[i-1, 'Holding']
            summary.loc[i, 'Cash'] = summary.loc[i-1, 'Cash']
            summary = update_row(summary, i)

        elif (summary.loc[i, '(VV - V) / VV'] + threshold > 0) & (summary.loc[i-1, 'Holding'] == 0): #buy
            #print(buy)
            summary.loc[i, 'Holding'] = (summary.loc[i-1, 'Cash'] - trade_cost) / summary.loc[i, 'SPY']
            summary.loc[i, 'Cash'] = 0
            summary = update_row(summary, i)

        elif (summary.loc[i, '(VV - V) / VV'] - threshold < 0) & (summary.loc[i-1, 'Held Value'] != 0): #sell
            #print('sell')
            summary.loc[i, 'Cash'] = summary.loc[i-1, 'Held Value'] - trade_cost
            summary.loc[i, 'Holding'] = 0
            summary.loc[i, 'Held Value'] = 0
            summary = update_row(summary, i)

        else: #also do nothing
            #print('do nothing')
            summary.loc[i, 'Holding'] = summary.loc[i-1, 'Holding']
            summary.loc[i, 'Cash'] = summary.loc[i-1, 'Cash']
            summary = update_row(summary, i)
            
    return summary
```

```python
#if -[((VV-V)/VV) - ((V7-V4)/(V7*3))] > 0 , buy SPY. If < 0, sell SPY

def simulate_returnsV47SPY(dataframe, initial_investment, trade_cost, threshold, start_date=None):
    """
    Simulates returns from buying SPY using existing dataset (fixed dataset based on summary dataframe)
    allows for varying the initial_investment, trade_cost, threshold, and start date (date must be in dataframe)
    Utilizes new formula [((VV-V)/VV) - ((V7-V4)/(V7*3))]
    """
    
    if start_date is not None:
        d1 = np.where(dataframe['Date'].values == start_date)[0][0]
        summary = dataframe.iloc[d1:, :].copy()
        summary.reset_index(drop = True, inplace = True)
    else:
        summary = dataframe.copy()
    
    summary = dataframe.copy()
    
    summary.loc[0, 'Holding':] = 0 #initialize as 0 holdings
    LTshares = (initial_investment - trade_cost) / summary.loc[0, 'SPY']
    summary.loc[0, 'Long Term Shares'] = LTshares
    summary.loc[0, 'Long Term Value'] = initial_investment - trade_cost
    summary.loc[0, 'Cash'] = initial_investment

    def update_row(df, row):
        """
        basic steps that will be repeated to update each row of summary DF
        """
        df.loc[row, 'Held Value'] = df.loc[row, 'Holding']*df.loc[row, 'SPY']
        df.loc[row, 'Long Term Shares'] = LTshares
        df.loc[row, 'Long Term Value'] = df.loc[row, 'Long Term Shares']*df.loc[row, 'SPY']
        df.loc[row, 'Trade Sum'] = df.loc[row, 'Cash'] + df.loc[row, 'Held Value']
        return df

    for i in range(1, summary.shape[0]):
        #print(i)
        if np.absolute(summary.loc[i, '[((VV-V)/VV) - ((V7-V4)/(V7*3))]']) <= threshold: #do nothing
            #print('do nothing')
            summary.loc[i, 'Holding'] = summary.loc[i-1, 'Holding']
            summary.loc[i, 'Cash'] = summary.loc[i-1, 'Cash']
            summary = update_row(summary, i)

        elif (summary.loc[i, '[((VV-V)/VV) - ((V7-V4)/(V7*3))]'] + threshold > 0) & (summary.loc[i-1, 'Holding'] == 0): #buy
            #print(buy)
            summary.loc[i, 'Holding'] = (summary.loc[i-1, 'Cash'] - trade_cost) / summary.loc[i, 'SPY']
            summary.loc[i, 'Cash'] = 0
            summary = update_row(summary, i)

        elif (summary.loc[i, '[((VV-V)/VV) - ((V7-V4)/(V7*3))]'] - threshold < 0) & (summary.loc[i-1, 'Held Value'] != 0): #sell
            #print('sell')
            summary.loc[i, 'Cash'] = summary.loc[i-1, 'Held Value'] - trade_cost
            summary.loc[i, 'Holding'] = 0
            summary.loc[i, 'Held Value'] = 0
            summary = update_row(summary, i)

        else: #also do nothing
            #print('do nothing')
            summary.loc[i, 'Holding'] = summary.loc[i-1, 'Holding']
            summary.loc[i, 'Cash'] = summary.loc[i-1, 'Cash']
            summary = update_row(summary, i)
            
    return summary
```

```python
# if -[((VV-V)/VV) - ((V7-V4)/(V7*3))] > 0, short VXX. If < 0, cover VXX

def simulate_returnsV47VXX(dataframe, initial_investment, trade_cost, threshold, start_date=None):
    """
    Simulates returns from shorting VXX using existing dataset (fixed dataset based on summary dataframe)
    allows for varying the initial_investment, trade_cost, threshold, and start date (date must be in dataframe)
    Utilizes new formula [((VV-V)/VV) - ((V7-V4)/(V7*3))]
    """
    
    if start_date is not None:
        d1 = np.where(dataframe['Date'].values == start_date)[0][0]
        summary = dataframe.iloc[d1:, :].copy()
        summary.reset_index(drop = True, inplace = True)
    else:
        summary = dataframe.copy()
    
    summary.loc[0, 'Holding':] = 0 #initialize as 0 holdings
    LTshares = (initial_investment - trade_cost) / summary.loc[0, 'SPY']
    summary.loc[0, 'Long Term Shares'] = LTshares
    summary.loc[0, 'Long Term Value'] = initial_investment - trade_cost
    summary.loc[0, 'Cash'] = initial_investment

    def update_row(df, row):
        """
        basic steps that will be repeated to update each row of summary DF
        """
        df.loc[row, 'Held Value'] = df.loc[row, 'Holding']*df.loc[row, 'VXX']
        df.loc[row, 'Long Term Shares'] = LTshares
        df.loc[row, 'Long Term Value'] = df.loc[row, 'Long Term Shares']*df.loc[row, 'SPY']
        df.loc[row, 'Trade Sum'] = df.loc[row, 'Cash'] + df.loc[row, 'Held Value']
        return df

    for i in range(1, summary.shape[0]):
        #print(i)
        if np.absolute(summary.loc[i, '[((VV-V)/VV) - ((V7-V4)/(V7*3))]']) <= threshold: #do nothing
            #print('do nothing')
            summary.loc[i, 'Holding'] = summary.loc[i-1, 'Holding']
            summary.loc[i, 'Cash'] = summary.loc[i-1, 'Cash']
            summary = update_row(summary, i)

        elif (summary.loc[i, '[((VV-V)/VV) - ((V7-V4)/(V7*3))]'] + threshold > 0) & (summary.loc[i-1, 'Holding'] == 0): #buy
            #print(short sell)
            summary.loc[i, 'Holding'] = (summary.loc[i-1, 'Cash'] - trade_cost) / -summary.loc[i, 'VXX']
            summary.loc[i, 'Cash'] = summary.loc[i-1, 'Cash'] - trade_cost -(summary.loc[i, 'Holding']*summary.loc[i, 'VXX'])
            summary = update_row(summary, i)

        elif (summary.loc[i, '[((VV-V)/VV) - ((V7-V4)/(V7*3))]'] - threshold < 0) & (summary.loc[i-1, 'Held Value'] != 0): #sell
            #print('sell')
            summary.loc[i, 'Cash'] = summary.loc[i-1, 'Cash'] + summary.loc[i-1, 'Held Value'] - trade_cost
            summary.loc[i, 'Holding'] = 0
            summary.loc[i, 'Held Value'] = 0
            summary = update_row(summary, i)

        else: #also do nothing
            #print('do nothing')
            summary.loc[i, 'Holding'] = summary.loc[i-1, 'Holding']
            summary.loc[i, 'Cash'] = summary.loc[i-1, 'Cash']
            summary = update_row(summary, i)
            
    return summary
```
